My_daily routine.

```python
# This code simulates my_daily morning routine,
# demonstrating the use of if, for, and while statements.

# --- Initial State Setup ---
current_time_minutes = 6 * 60  # Starting at 6:00 AM (360 minutes)
LEAVING_TIME_MINUTES = 8 * 60  # Goal: Leave for school by 8:00 AM (480 minutes)
is_ready_for_school = False
is_prayer_day = True # A flag to simulate a decision based on the day
books_to_pack = ["Computer Textbook", "All_today's course units", "Laptop", "Water Bottle"]

print("--- My_Morning Routine Initiated ---")
# Using a function to keep the time display clean
def format_time(minutes):
    """Converts total minutes into HH:MM AM/PM format."""
    hour = minutes // 60
    minute = minutes % 60
    am_pm = "AM"
    if hour >= 12:
        am_pm = "PM"
        if hour > 12:
            hour -= 12
    return f"{hour:02d}:{minute:02d} {am_pm}"

# --- 1. Waking Up (IF/ELIF) ---

# Decision upon waking up
if current_time_minutes < 6 * 60:
    # This handles if the code somehow started too early
    print(f"It's only {format_time(current_time_minutes)}. Going back to sleep for a bit...")
    current_time_minutes = 6 * 60
elif current_time_minutes == 6 * 60:
    print(f"Alarm set for {format_time(current_time_minutes)}! Time to get up.")

# Advance time for quick preparation.
current_time_minutes += 30 # Time is now 6:30 AM
print(f"-> Quick shower and dressed. Time is now {format_time(current_time_minutes)}")


# --- 2. Morning Activities Loop (WHILE) ---

# The WHILE loop controls the main routine, continuing as long as:
# 1. I am NOT ready to leave, AND
# 2. They haven't passed the critical leaving time.
while not is_ready_for_school and current_time_minutes < LEAVING_TIME_MINUTES:
    
    time_remaining = LEAVING_TIME_MINUTES - current_time_minutes
    print("-" * 20)
    print(f"Current time: {format_time(current_time_minutes)}. {time_remaining} minutes remaining.")
    
    # --- Nested IF/ELIF: Decision making based on schedule/time ---
    
    if is_prayer_day:
        # Check if there is enough time for the scheduled activity.
        if time_remaining >= 60:
            print("Decision: It's a prayer day and I have enough time (>= 60 mins). Heading out for a short prayer!")
            current_time_minutes += 60 # 1 hour prayer.
            print(f"-> Finished prayer session. Time is now {format_time(current_time_minutes)}")
            is_prayer_day = False # Activity completed
        else:
            print("Decision: Not enough time left for the prayer. Skipping today.")
            is_prayer_day = False
    
    elif current_time_minutes < 7 * 60 + 15: # Before 7:15 AM.
        print("Decision: Time for a focused 45-minute study session before breakfast.")
        
        # --- FOR Loop: Simulating reviewing specific study topics ---
        study_topics = ["Python essentials", "Data Science & Analysis", "Assignment Deadlines"]
        for i, topic in enumerate(study_topics):
            print(f"  -> Reviewing topic {i+1}: {topic}")
            current_time_minutes += 15 # 15 minutes per topic review.
        
        print(f"-> Study session complete. Time is now {format_time(current_time_minutes)}")

    elif current_time_minutes < 7 * 60 + 40: # Between 7:15 AM and 7:40 AM
        print("Decision: Preparing and eating a quick breakfast.")
        
        # --- FOR Loop: Simulating consumption of breakfast items ---
        breakfast_items = ["Oatmeal", "Coffee", "Banana"]
        for item in breakfast_items:
            print(f"  -> Consuming: {item}")
            current_time_minutes += 5 # 5 minutes per item.
            
        print(f"-> Breakfast done. Time is now {format_time(current_time_minutes)}")
        
    else:
        # If previous major activity slots are missed or finished, focus on packing
        print("Decision: Final check and packing bags.")
        
        # --- FOR Loop: Simulating packing the essential items ---
        print("Packing routine:")
        for item in books_to_pack:
            print(f"  -> Packed: {item}")
            current_time_minutes += 2 # 2 minutes per item
        
        is_ready_for_school = True # Set the flag to exit the WHILE loop

# --- 3. Final Check and Departure (IF/ELIF) ---

print("\n" + "=" * 30)

# Final check using IF/ELIF to determine if the departure was successful/timely
if is_ready_for_school and current_time_minutes <= LEAVING_TIME_MINUTES:
    print(f"Success! I am ready and leaving for school at {format_time(current_time_minutes)}.")
    print("I have plenty of time to catch the Bodaboda and grab a coffee.")
elif is_ready_for_school and current_time_minutes > LEAVING_TIME_MINUTES:
    print(f"Oh no! I was ready, but I left late at {format_time(current_time_minutes)}.")
    print("I will have to rush to make it to my first class!")
else:
    # This executes if the loop ended because time ran out, but I wasn't fully ready.
    print(f"Time's up! The current time is {format_time(current_time_minutes)}. I must leave immediately, even if I'm not fully ready!")

print("--- End of Morning Routine ---")
```
